```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate   
import seaborn as sns
```

```python
df=pd.read_csv('AB_NYC_2019.csv')
df
```

```python
df.describe()            # Summary statistics of the dataset
```

```python
df.isnull().sum()               #count of Missing values in dataset
```

```python
df.info()                    # Count of non null values in dataset
```

```python
def get_column_datatype(column):
    if df[column].dtype=="object":
        return "categorical"
    else:
        return "numerical"
categorical_columns=[]
numerical_columns=[]
for column in df.columns:
    column_datatype=get_column_datatype(column)
    if column_datatype=="categorical":
        categorical_columns.append(column)
    else:
        numerical_columns.append(column)
print("Categorical Columns:")        
print(categorical_columns)
print("\nNumerical Columns:")        
print(numerical_columns)
```

```python
#replacing missing values by mode
name_mode = df.name.mode()[0]
df["name"].fillna(name_mode, inplace = True)
```

```python
last_review_mode=df.last_review.mode()[0]
df["last_review"].fillna(last_review_mode, inplace = True)
```

```python
# dropping rows where host name is not provided
df=df.dropna(subset=['host_name'])
```

```python
#imputing mean in numerical columns with missing values
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
#box plot for numerical columns
for column in numerical_columns:
    sns.boxplot(data=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
#Removing outliers by IQR method
```

```python
columns_with_outliers=['host_id','latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count']
for column in columns_with_outliers:
    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR= Q3-Q1
    lower_bound= Q1-1.5*IQR
    upper_bound= Q3+1.5*IQR
    
    df=df[(df[column]>=lower_bound) & (df[column]<=upper_bound)]
```
